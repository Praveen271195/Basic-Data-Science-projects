## K Means Cluster

```python
# Import KMeans from SciKit Learn
from sklearn.cluster import KMeans
```

```python
#  Creating an instance of a K Means model with 2 clusters 
Kmodel = KMeans(n_clusters=2)
```

```python
# Fitting the model to all the data except for the Private label
final_data = data.drop("Private",axis=1)
final_data = final_data.drop("Unnamed: 0",axis=1)
```

```python
Kmodel.fit(final_data)
```

```python
# Cluster center vectors
Kmodel.cluster_centers_
```

## Evaluation

```python
# Performing one hot encoding on the "private column"
a = pd.DataFrame(data["Private"])
b = pd.get_dummies(a,drop_first=True)
```

```python
# Concatinating data and b
data = pd.concat([data,b],axis=1)
```

```python
# Checking 
data.head()
```

```python
# Importing the metrics for evaluation
from sklearn.metrics import confusion_matrix,classification_report
```

```python
# Scores
print("Confussion Matrix:")
print(confusion_matrix(data["Private_Yes"],Kmodel.labels_))
print("\n")
print("Classification Report:")
print(classification_report(data["Private_Yes"],Kmodel.labels_))
```
